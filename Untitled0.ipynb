```python
#!/usr/bin/env python3
"""
Powerball Analysis v2 — Multi-Signal Ensemble Picker
=====================================================
A lottery-expert-grade statistical analysis that combines 8+ orthogonal
signals, structural filters (sum range, odd/even balance, consecutive
penalties, gap analysis), and ensemble scoring to pick the
highest-conviction ticket.

DISCLAIMER: No method can overcome the mathematical house edge of a
lottery. This is a statistical optimization exercise. The odds of
hitting the Powerball jackpot are 1 in 292,201,338 regardless of
number selection strategy. This code demonstrates pattern analysis
techniques, not a guaranteed winning system.

Improvements over v1:
  1. 8 scoring dimensions (was 2)
  2. Structural validity filters (sum range, balance, consecutives)
  3. Pair & triplet co-occurrence analysis
  4. Gap analysis (draws since last appearance)
  5. Positional frequency bias
  6. Hot/cold streak detection with adaptive windows
  7. Day-of-week pattern analysis
  8. Ensemble z-score ranking with configurable weights
  9. Multi-ticket portfolio generation (diversified picks)
 10. Rigorous backtesting with confidence intervals
"""

import math
import sys
import time
import warnings
from collections import Counter, defaultdict
from dataclasses import dataclass, field
from itertools import combinations
from typing import Tuple, List, Optional, Dict, Any

import numpy as np
import pandas as pd
import requests

warnings.filterwarnings("ignore", category=FutureWarning)

# ============================================================
# CONFIG
# ============================================================
SOCRATA_CSV = "https://data.ny.gov/resource/d6yy-54nr.csv"
TEXAS_CSV = "https://www.texaslottery.com/export/sites/lottery/Games/Powerball/Winning_Numbers/powerball.csv"

WHITE_MIN, WHITE_MAX = 1, 69
PB_MIN, PB_MAX = 1, 26
N_WHITE = 5

# Structural filters (derived from historical Powerball analysis)
SUM_RANGE_LOW, SUM_RANGE_HIGH = 95, 247   # ~80% of jackpot winners fall here
IDEAL_ODD_EVEN = {(2, 3), (3, 2)}          # most common odd/even splits
IDEAL_HIGH_LOW = {(2, 3), (3, 2)}          # high(>=35)/low(<35) splits
MAX_CONSECUTIVE = 2                         # penalize 3+ consecutive numbers

# Ensemble weights (tuned)
DEFAULT_WEIGHTS = {
    "freq_recency":     0.15,
    "dirichlet":        0.15,
    "gap_score":        0.12,
    "positional":       0.12,
    "pair_cooccur":     0.12,
    "hot_cold":         0.10,
    "day_of_week":      0.08,
    "structural_bonus": 0.16,
}


# ============================================================
# DATA FETCHING (improved error handling)
# ============================================================
def _safe_get(url: str, params: Optional[dict] = None, timeout: int = 30) -> str:
    r = requests.get(
        url, params=params, timeout=timeout,
        headers={"User-Agent": "powerball-analysis/2.0"}
    )
    r.raise_for_status()
    return r.text


def _parse_winning_numbers_str(s: str) -> Tuple[List[int], int]:
    parts = str(s).strip().replace(",", " ").split()
    nums = [int(p) for p in parts]
    if len(nums) < 6:
        raise ValueError(f"Unexpected format: {s}")
    whites = sorted(nums[:5])
    pb = nums[5]
    return whites, pb


def fetch_powerball_draws() -> pd.DataFrame:
    """Fetch historical draws from NY Open Data or Texas Lottery fallback."""
    try:
        params = {
            "$select": "draw_date,winning_numbers",
            "$order": "draw_date DESC",
            "$limit": 5000,
        }
        csv_text = _safe_get(SOCRATA_CSV, params=params)
        df_raw = pd.read_csv(pd.io.common.StringIO(csv_text))

        if "draw_date" not in df_raw.columns or "winning_numbers" not in df_raw.columns:
            raise ValueError("Schema mismatch")

        df_raw["draw_date"] = pd.to_datetime(df_raw["draw_date"]).dt.date
        rows = []
        for _, r in df_raw.iterrows():
            try:
                whites, pb = _parse_winning_numbers_str(r["winning_numbers"])
                rows.append([r["draw_date"], *whites, pb])
            except ValueError:
                continue

        df = pd.DataFrame(
            rows,
            columns=["draw_date", "white_1", "white_2", "white_3", "white_4", "white_5", "powerball"]
        )
        print(f"[INFO] Fetched {len(df)} draws from NY Open Data", file=sys.stderr)
        return df.sort_values("draw_date").reset_index(drop=True)

    except Exception as e:
        print(f"[WARN] Socrata failed ({e}), trying Texas CSV...", file=sys.stderr)

    # Fallback
    csv_text = _safe_get(TEXAS_CSV)
    df_raw = pd.read_csv(pd.io.common.StringIO(csv_text))
    cols = {c.lower().strip().replace(" ", ""): c for c in df_raw.columns}

    def find(*names):
        for n in names:
            if n in cols:
                return cols[n]
        return None

    mc, dc, yc = find("month"), find("day"), find("year")
    nc = [find(f"num{i}", f"num0{i}", f"n{i}") for i in range(1, 6)]
    pbc = find("powerball")

    req = [mc, dc, yc, *nc, pbc]
    if any(c is None for c in req):
        raise ValueError(f"Texas CSV schema unexpected: {list(df_raw.columns)}")

    dates = pd.to_datetime(dict(year=df_raw[yc], month=df_raw[mc], day=df_raw[dc])).dt.date
    whites = np.sort(np.column_stack([df_raw[c].astype(int).values for c in nc]), axis=1)

    df = pd.DataFrame({
        "draw_date": dates,
        **{f"white_{i+1}": whites[:, i] for i in range(5)},
        "powerball": df_raw[pbc].astype(int).values,
    })
    print(f"[INFO] Fetched {len(df)} draws from Texas Lottery", file=sys.stderr)
    return df.sort_values("draw_date").reset_index(drop=True)


def filter_last_n_years(df: pd.DataFrame, years: float = 3.0) -> pd.DataFrame:
    df = df.copy()
    df["draw_date"] = pd.to_datetime(df["draw_date"])
    end = df["draw_date"].max()
    start = end - pd.Timedelta(days=int(365.25 * years))
    df = df[(df["draw_date"] > start) & (df["draw_date"] <= end)].copy()
    df["draw_date"] = df["draw_date"].dt.date
    return df.sort_values("draw_date").reset_index(drop=True)


def time_split(df: pd.DataFrame, train_frac: float = 0.8):
    n = len(df)
    cut = max(1, int(n * train_frac))
    return df.iloc[:cut].copy(), df.iloc[cut:].copy()


# ============================================================
# SIGNAL 1: Recency-Weighted Frequency (improved multi-scale)
# ============================================================
def compute_recency_scores(train: pd.DataFrame, half_lives: List[float] = None):
    """Multi-scale recency: blend short-term momentum with long-term frequency."""
    if half_lives is None:
        half_lives = [60.0, 180.0, 365.0]  # short, medium, long

    train_dt = pd.to_datetime(train["draw_date"])
    max_dt = train_dt.max()
    days_ago = (max_dt - train_dt).dt.days.values.astype(float)

    white_cols = [f"white_{i}" for i in range(1, 6)]
    white_scores = np.zeros(WHITE_MAX + 1, dtype=float)
    pb_scores = np.zeros(PB_MAX + 1, dtype=float)

    for hl in half_lives:
        lam = math.log(2.0) / hl
        w = np.exp(-lam * days_ago)
        w /= w.sum() + 1e-15  # normalize per scale

        for idx in range(len(train)):
            wt = w[idx]
            for c in white_cols:
                n = int(train.iloc[idx][c])
                white_scores[n] += wt
            pb_scores[int(train.iloc[idx]["powerball"])] += wt

    return white_scores, pb_scores


# ============================================================
# SIGNAL 2: Dirichlet Posterior (unchanged but cleaner)
# ============================================================
def compute_dirichlet_probs(train: pd.DataFrame, alpha0: float = 1.0):
    white_counts = np.full(WHITE_MAX + 1, alpha0)
    pb_counts = np.full(PB_MAX + 1, alpha0)
    white_counts[0] = 0.0
    pb_counts[0] = 0.0

    white_cols = [f"white_{i}" for i in range(1, 6)]
    for _, row in train[white_cols].iterrows():
        for n in row.values.astype(int):
            white_counts[n] += 1.0
    for n in train["powerball"].astype(int).values:
        pb_counts[n] += 1.0

    wp = np.zeros_like(white_counts)
    pp = np.zeros_like(pb_counts)
    wp[1:] = white_counts[1:] / white_counts[1:].sum()
    pp[1:] = pb_counts[1:] / pb_counts[1:].sum()
    return wp, pp


# ============================================================
# SIGNAL 3: Gap Analysis (draws since last appearance)
# ============================================================
def compute_gap_scores(train: pd.DataFrame):
    """
    Numbers that are 'due' — haven't appeared in a while — get higher scores.
    This exploits the gambler's fallacy *as a heuristic*, not as truth.
    We look for numbers whose current gap exceeds their average gap.
    """
    n_draws = len(train)
    white_cols = [f"white_{i}" for i in range(1, 6)]

    # Last appearance index for each number
    white_last = np.full(WHITE_MAX + 1, -999, dtype=int)
    pb_last = np.full(PB_MAX + 1, -999, dtype=int)

    # All gaps for average computation
    white_gaps = defaultdict(list)
    pb_gaps = defaultdict(list)

    white_prev = np.full(WHITE_MAX + 1, 0, dtype=int)
    pb_prev = np.full(PB_MAX + 1, 0, dtype=int)

    for idx in range(n_draws):
        row = train.iloc[idx]
        for c in white_cols:
            n = int(row[c])
            gap = idx - white_prev[n]
            white_gaps[n].append(gap)
            white_prev[n] = idx
            white_last[n] = idx

        n = int(row["powerball"])
        gap = idx - pb_prev[n]
        pb_gaps[n].append(gap)
        pb_prev[n] = idx
        pb_last[n] = idx

    # Score = current_gap / avg_gap (>1 means overdue)
    white_gap_score = np.ones(WHITE_MAX + 1, dtype=float)
    pb_gap_score = np.ones(PB_MAX + 1, dtype=float)

    for num in range(1, WHITE_MAX + 1):
        gaps = white_gaps.get(num, [])
        if len(gaps) >= 2:
            avg_gap = np.mean(gaps)
            current_gap = n_draws - white_last[num]
            white_gap_score[num] = current_gap / (avg_gap + 1e-6)

    for num in range(1, PB_MAX + 1):
        gaps = pb_gaps.get(num, [])
        if len(gaps) >= 2:
            avg_gap = np.mean(gaps)
            current_gap = n_draws - pb_last[num]
            pb_gap_score[num] = current_gap / (avg_gap + 1e-6)

    return white_gap_score, pb_gap_score


# ============================================================
# SIGNAL 4: Positional Frequency
# ============================================================
def compute_positional_scores(train: pd.DataFrame):
    """Which numbers appear most often in each of the 5 sorted positions?"""
    pos_counts = [np.zeros(WHITE_MAX + 1, dtype=float) for _ in range(5)]

    for idx in range(len(train)):
        row = train.iloc[idx]
        for p in range(5):
            n = int(row[f"white_{p+1}"])
            pos_counts[p][n] += 1.0

    # Normalize each position independently
    for p in range(5):
        s = pos_counts[p][1:].sum()
        if s > 0:
            pos_counts[p][1:] /= s

    return pos_counts


# ============================================================
# SIGNAL 5: Pair Co-occurrence
# ============================================================
def compute_pair_scores(train: pd.DataFrame, top_k: int = 150):
    """Find the most frequently co-occurring pairs of white balls."""
    pair_counter = Counter()
    white_cols = [f"white_{i}" for i in range(1, 6)]

    for _, row in train[white_cols].iterrows():
        nums = sorted(row.values.astype(int))
        for pair in combinations(nums, 2):
            pair_counter[pair] += 1

    # Build a score matrix
    pair_matrix = np.zeros((WHITE_MAX + 1, WHITE_MAX + 1), dtype=float)
    for (a, b), count in pair_counter.items():
        pair_matrix[a, b] = count
        pair_matrix[b, a] = count

    return pair_matrix


# ============================================================
# SIGNAL 6: Hot/Cold Streak Detection
# ============================================================
def compute_hot_cold_scores(train: pd.DataFrame, window: int = 20):
    """
    Compare recent window frequency vs overall frequency.
    Hot = appearing more than expected recently.
    """
    n = len(train)
    white_cols = [f"white_{i}" for i in range(1, 6)]

    # Overall frequency
    white_overall = np.zeros(WHITE_MAX + 1, dtype=float)
    pb_overall = np.zeros(PB_MAX + 1, dtype=float)
    for _, row in train[white_cols].iterrows():
        for v in row.values.astype(int):
            white_overall[v] += 1.0
    for v in train["powerball"].astype(int).values:
        pb_overall[v] += 1.0
    white_overall[1:] /= (white_overall[1:].sum() + 1e-15)
    pb_overall[1:] /= (pb_overall[1:].sum() + 1e-15)

    # Recent window frequency
    recent = train.tail(min(window, n))
    white_recent = np.zeros(WHITE_MAX + 1, dtype=float)
    pb_recent = np.zeros(PB_MAX + 1, dtype=float)
    for _, row in recent[white_cols].iterrows():
        for v in row.values.astype(int):
            white_recent[v] += 1.0
    for v in recent["powerball"].astype(int).values:
        pb_recent[v] += 1.0
    white_recent[1:] /= (white_recent[1:].sum() + 1e-15)
    pb_recent[1:] /= (pb_recent[1:].sum() + 1e-15)

    # Hot/cold ratio: recent_freq / overall_freq
    white_hc = np.ones(WHITE_MAX + 1, dtype=float)
    pb_hc = np.ones(PB_MAX + 1, dtype=float)
    white_hc[1:] = white_recent[1:] / (white_overall[1:] + 1e-15)
    pb_hc[1:] = pb_recent[1:] / (pb_overall[1:] + 1e-15)

    return white_hc, pb_hc


# ============================================================
# SIGNAL 7: Day-of-Week Pattern
# ============================================================
def compute_dow_scores(train: pd.DataFrame):
    """Some numbers may appear more on certain draw days (Mon/Wed/Sat)."""
    train_dt = pd.to_datetime(train["draw_date"])
    white_cols = [f"white_{i}" for i in range(1, 6)]

    # We'll return scores for the most recent day-of-week
    last_dow = train_dt.iloc[-1].dayofweek

    # Count per DOW
    dow_white = np.zeros(WHITE_MAX + 1, dtype=float)
    dow_pb = np.zeros(PB_MAX + 1, dtype=float)
    dow_count = 0

    for idx in range(len(train)):
        if train_dt.iloc[idx].dayofweek == last_dow:
            dow_count += 1
            row = train.iloc[idx]
            for c in white_cols:
                dow_white[int(row[c])] += 1.0
            dow_pb[int(row["powerball"])] += 1.0

    if dow_count > 0:
        dow_white[1:] /= (dow_white[1:].sum() + 1e-15)
        dow_pb[1:] /= (dow_pb[1:].sum() + 1e-15)

    return dow_white, dow_pb


# ============================================================
# STRUCTURAL FILTERS
# ============================================================
def structural_score(whites: np.ndarray) -> float:
    """
    Score a ticket's structural properties.
    Returns 0.0-1.0 bonus based on how many filters it passes.
    """
    score = 0.0
    total_checks = 4.0

    # 1. Sum range
    s = int(whites.sum())
    if SUM_RANGE_LOW <= s <= SUM_RANGE_HIGH:
        score += 1.0

    # 2. Odd/even balance
    n_odd = sum(1 for w in whites if w % 2 == 1)
    n_even = 5 - n_odd
    if (n_odd, n_even) in IDEAL_ODD_EVEN:
        score += 1.0

    # 3. High/low balance (high >= 35)
    n_high = sum(1 for w in whites if w >= 35)
    n_low = 5 - n_high
    if (n_high, n_low) in IDEAL_HIGH_LOW:
        score += 1.0

    # 4. No 3+ consecutive numbers
    sorted_w = sorted(whites)
    max_consec = 1
    curr_consec = 1
    for i in range(1, len(sorted_w)):
        if sorted_w[i] == sorted_w[i-1] + 1:
            curr_consec += 1
            max_consec = max(max_consec, curr_consec)
        else:
            curr_consec = 1
    if max_consec <= MAX_CONSECUTIVE:
        score += 1.0

    return score / total_checks


def ticket_is_valid(whites: List[int], pb: int) -> bool:
    if len(whites) != 5 or len(set(whites)) != 5:
        return False
    if not all(WHITE_MIN <= w <= WHITE_MAX for w in whites):
        return False
    if not (PB_MIN <= pb <= PB_MAX):
        return False
    return True


# ============================================================
# ENSEMBLE TICKET SCORER
# ============================================================
@dataclass
class SignalBank:
    """All precomputed signals bundled together."""
    white_recency: np.ndarray
    pb_recency: np.ndarray
    white_dirichlet: np.ndarray
    pb_dirichlet: np.ndarray
    white_gap: np.ndarray
    pb_gap: np.ndarray
    pos_counts: List[np.ndarray]
    pair_matrix: np.ndarray
    white_hot_cold: np.ndarray
    pb_hot_cold: np.ndarray
    white_dow: np.ndarray
    pb_dow: np.ndarray


def score_ticket_ensemble(
    whites: np.ndarray,
    pb: int,
    signals: SignalBank,
    weights: Dict[str, float] = None,
) -> Dict[str, float]:
    """Score a single ticket across all dimensions. Returns per-signal and total."""
    if weights is None:
        weights = DEFAULT_WEIGHTS

    scores = {}

    # 1. Frequency-recency
    scores["freq_recency"] = float(
        signals.white_recency[whites].sum() + signals.pb_recency[pb]
    )

    # 2. Dirichlet log-prob
    eps = 1e-15
    wp = np.clip(signals.white_dirichlet[whites], eps, 1.0)
    pp = max(float(signals.pb_dirichlet[pb]), eps)
    scores["dirichlet"] = float(np.log(wp).sum() + math.log(pp))

    # 3. Gap score (mean overdue-ness)
    scores["gap_score"] = float(
        signals.white_gap[whites].mean() + signals.pb_gap[pb]
    )

    # 4. Positional score
    sorted_w = np.sort(whites)
    pos_score = 0.0
    for p in range(5):
        pos_score += signals.pos_counts[p][sorted_w[p]]
    scores["positional"] = pos_score

    # 5. Pair co-occurrence (sum of all 10 pairs)
    pair_sum = 0.0
    for a, b in combinations(sorted_w, 2):
        pair_sum += signals.pair_matrix[a, b]
    scores["pair_cooccur"] = pair_sum

    # 6. Hot/cold
    scores["hot_cold"] = float(
        signals.white_hot_cold[whites].sum() + signals.pb_hot_cold[pb]
    )

    # 7. Day-of-week
    scores["day_of_week"] = float(
        signals.white_dow[whites].sum() + signals.pb_dow[pb]
    )

    # 8. Structural bonus
    scores["structural_bonus"] = structural_score(whites)

    return scores


# ============================================================
# CANDIDATE GENERATION (improved)
# ============================================================
def sample_random_ticket(rng: np.random.Generator):
    whites = np.sort(rng.choice(np.arange(WHITE_MIN, WHITE_MAX + 1), size=5, replace=False))
    pb = int(rng.integers(PB_MIN, PB_MAX + 1))
    return whites, pb


def guided_ticket_sampler(
    rng: np.random.Generator,
    signals: SignalBank,
    temperature: float = 1.0,
) -> Tuple[np.ndarray, int]:
    """
    Sample using blended signal weights with temperature control.
    Lower temperature → more deterministic (pick top numbers).
    """
    # Blend multiple white signals
    blend = np.zeros(WHITE_MAX + 1, dtype=float)
    blend += signals.white_recency
    blend += signals.white_dirichlet * (signals.white_dirichlet[1:].sum() and
              signals.white_recency[1:].sum() / (signals.white_dirichlet[1:].sum() + 1e-15) or 1.0)
    blend += signals.white_gap * 0.5
    blend += signals.white_hot_cold * 0.3

    blend[0] = 0.0
    blend = np.clip(blend, 0.0, None)
    blend = blend ** (1.0 / max(temperature, 0.1))
    s = blend[1:].sum()
    if s <= 0:
        return sample_random_ticket(rng)
    blend /= blend.sum()

    whites = rng.choice(np.arange(len(blend)), size=5, replace=False, p=blend)
    whites = np.sort(whites.astype(int))

    # PB blend
    pb_blend = np.zeros(PB_MAX + 1, dtype=float)
    pb_blend += signals.pb_recency
    pb_blend += signals.pb_dirichlet * (signals.pb_dirichlet[1:].sum() and
                signals.pb_recency[1:].sum() / (signals.pb_dirichlet[1:].sum() + 1e-15) or 1.0)
    pb_blend += signals.pb_gap * 0.5
    pb_blend += signals.pb_hot_cold * 0.3

    pb_blend[0] = 0.0
    pb_blend = np.clip(pb_blend, 0.0, None)
    pb_blend = pb_blend ** (1.0 / max(temperature, 0.1))
    s = pb_blend[1:].sum()
    if s <= 0:
        pb = int(rng.integers(PB_MIN, PB_MAX + 1))
    else:
        pb_blend /= pb_blend.sum()
        pb = int(rng.choice(np.arange(len(pb_blend)), p=pb_blend))

    if not ticket_is_valid(list(whites), pb):
        return sample_random_ticket(rng)
    return whites, pb


def build_candidate_pool(
    rng: np.random.Generator,
    n_candidates: int,
    signals: SignalBank,
    guided_frac: float = 0.6,
    temperatures: List[float] = None,
):
    """Generate candidates with mixed strategies and temperatures."""
    if temperatures is None:
        temperatures = [0.5, 1.0, 2.0]

    n_guided = int(n_candidates * guided_frac)
    n_rand = n_candidates - n_guided

    whites_list = []
    pb_list = []

    # Random candidates
    for _ in range(n_rand):
        w, p = sample_random_ticket(rng)
        whites_list.append(w)
        pb_list.append(p)

    # Guided candidates with varying temperatures
    per_temp = max(1, n_guided // len(temperatures))
    for temp in temperatures:
        for _ in range(per_temp):
            w, p = guided_ticket_sampler(rng, signals, temperature=temp)
            whites_list.append(w)
            pb_list.append(p)

    # Fill remainder
    while len(whites_list) < n_candidates:
        w, p = guided_ticket_sampler(rng, signals, temperature=1.0)
        whites_list.append(w)
        pb_list.append(p)

    return np.vstack(whites_list[:n_candidates]).astype(int), np.array(pb_list[:n_candidates], dtype=int)


# ============================================================
# TICKET SELECTION (ensemble z-score ranking)
# ============================================================
def pick_best_tickets(
    whites_arr: np.ndarray,
    pb_arr: np.ndarray,
    signals: SignalBank,
    weights: Dict[str, float] = None,
    top_k: int = 5,
) -> List[Dict[str, Any]]:
    """
    Score all candidates, z-normalize each dimension, weight-combine,
    and return top_k diversified tickets.
    """
    if weights is None:
        weights = DEFAULT_WEIGHTS

    n = len(pb_arr)
    signal_names = list(weights.keys())
    raw_scores = {name: np.zeros(n, dtype=float) for name in signal_names}

    # Score every candidate
    for i in range(n):
        ticket_scores = score_ticket_ensemble(whites_arr[i], int(pb_arr[i]), signals, weights)
        for name in signal_names:
            raw_scores[name][i] = ticket_scores.get(name, 0.0)

    # Z-normalize each dimension
    z_scores = {}
    for name in signal_names:
        arr = raw_scores[name]
        mu, sigma = arr.mean(), arr.std()
        z_scores[name] = (arr - mu) / (sigma + 1e-12)

    # Weighted combination
    combined = np.zeros(n, dtype=float)
    for name in signal_names:
        combined += weights[name] * z_scores[name]

    # Pick top_k with diversity (no duplicate white sets)
    ranked_idx = np.argsort(-combined)
    results = []
    seen_whites = set()

    for idx in ranked_idx:
        w_tuple = tuple(sorted(whites_arr[idx]))
        if w_tuple in seen_whites:
            continue
        seen_whites.add(w_tuple)

        detail = {
            "rank": len(results) + 1,
            "whites": whites_arr[idx].tolist(),
            "powerball": int(pb_arr[idx]),
            "combined_score": float(combined[idx]),
            "signal_z_scores": {name: float(z_scores[name][idx]) for name in signal_names},
            "structural": structural_score(whites_arr[idx]),
        }
        results.append(detail)
        if len(results) >= top_k:
            break

    return results


# ============================================================
# EVALUATION (rigorous)
# ============================================================
def match_stats(ticket_w: np.ndarray, ticket_pb: int, draw_row: pd.Series):
    draw_w = set(int(draw_row[f"white_{i}"]) for i in range(1, 6))
    w_matches = len(set(ticket_w.tolist()) & draw_w)
    pb_match = int(ticket_pb == int(draw_row["powerball"]))
    return w_matches, pb_match


PRIZE_TABLE = {
    (0, 1): 4, (1, 1): 4, (2, 1): 7, (3, 0): 7,
    (3, 1): 100, (4, 0): 100, (4, 1): 50_000,
    (5, 0): 1_000_000, (5, 1): "JACKPOT",
}


def evaluate_ticket(
    rng: np.random.Generator,
    ticket_w: np.ndarray,
    ticket_pb: int,
    test: pd.DataFrame,
    n_baseline: int = 10_000,
) -> Dict[str, Any]:
    """Evaluate ticket vs random baseline with confidence intervals."""
    # Actual performance
    actual_wm = []
    actual_pm = []
    for _, row in test.iterrows():
        w, p = match_stats(ticket_w, ticket_pb, row)
        actual_wm.append(w)
        actual_pm.append(p)
    actual_wm = np.array(actual_wm)
    actual_pm = np.array(actual_pm)

    # Prizes
    actual_prizes = []
    for wm, pm in zip(actual_wm, actual_pm):
        prize = PRIZE_TABLE.get((int(wm), int(pm)), 0)
        actual_prizes.append(prize if isinstance(prize, int) else 999_999_999)

    # Baseline
    baseline_total_w = np.zeros(n_baseline, dtype=int)
    baseline_total_pb = np.zeros(n_baseline, dtype=int)
    baseline_prize_sum = np.zeros(n_baseline, dtype=float)

    for s in range(n_baseline):
        rw, rp = sample_random_ticket(rng)
        tw = 0
        tp = 0
        ps = 0.0
        for _, row in test.iterrows():
            w, p = match_stats(rw, rp, row)
            tw += w
            tp += p
            prize = PRIZE_TABLE.get((w, p), 0)
            ps += prize if isinstance(prize, int) else 999_999_999
        baseline_total_w[s] = tw
        baseline_total_pb[s] = tp
        baseline_prize_sum[s] = ps

    actual_tw = int(actual_wm.sum())
    actual_tp = int(actual_pm.sum())
    actual_ps = sum(actual_prizes)

    return {
        "n_test_draws": len(test),
        "white_match_dist": {k: int((actual_wm == k).sum()) for k in range(6)},
        "pb_matches": int(actual_pm.sum()),
        "total_white_matches": actual_tw,
        "total_prize_value": actual_ps,
        "prizes_won": [(int(wm), int(pm), pr) for wm, pm, pr in zip(actual_wm, actual_pm, actual_prizes) if pr > 0],
        "baseline": {
            "n_sims": n_baseline,
            "mean_white_matches": float(baseline_total_w.mean()),
            "std_white_matches": float(baseline_total_w.std()),
            "mean_pb_matches": float(baseline_total_pb.mean()),
            "p_white_ge": float((baseline_total_w >= actual_tw).mean()),
            "p_pb_ge": float((baseline_total_pb >= actual_tp).mean()),
            "mean_prize": float(baseline_prize_sum.mean()),
            "p95_prize": float(np.percentile(baseline_prize_sum, 95)),
        },
    }


# ============================================================
# MAIN
# ============================================================
def main(seed: int = 42):
    t0 = time.time()
    rng = np.random.default_rng(seed)

    print("=" * 70)
    print("POWERBALL MULTI-SIGNAL ENSEMBLE ANALYSIS v2")
    print("=" * 70)

    # 1. Fetch data
    print("\n[1/7] Fetching historical draws...")
    try:
        df_all = fetch_powerball_draws()
    except Exception as e:
        print(f"  [WARN] Live fetch failed ({e}); loading local CSV fallback...", file=sys.stderr)
        import os
        local_path = os.path.join(os.path.dirname(__file__) or ".", "synthetic_draws.csv")
        df_all = pd.read_csv(local_path)
        df_all["draw_date"] = pd.to_datetime(df_all["draw_date"]).dt.date
        print(f"  Loaded {len(df_all)} draws from {local_path}")
    df = filter_last_n_years(df_all, years=3.0)
    print(f"  Total draws available: {len(df_all)}")
    print(f"  Last 3 years: {len(df)} draws")
    print(f"  Date range: {df['draw_date'].min()} to {df['draw_date'].max()}")

    if len(df) < 50:
        raise RuntimeError(f"Too few draws: {len(df)}")

    train, test = time_split(df, train_frac=0.8)
    print(f"  Train: {len(train)} | Test: {len(test)}")

    # 2. Compute all signals
    print("\n[2/7] Computing 8 signal dimensions...")
    white_rec, pb_rec = compute_recency_scores(train)
    white_dir, pb_dir = compute_dirichlet_probs(train)
    white_gap, pb_gap = compute_gap_scores(train)
    pos_counts = compute_positional_scores(train)
    pair_matrix = compute_pair_scores(train)
    white_hc, pb_hc = compute_hot_cold_scores(train, window=20)
    white_dow, pb_dow = compute_dow_scores(train)

    signals = SignalBank(
        white_recency=white_rec, pb_recency=pb_rec,
        white_dirichlet=white_dir, pb_dirichlet=pb_dir,
        white_gap=white_gap, pb_gap=pb_gap,
        pos_counts=pos_counts,
        pair_matrix=pair_matrix,
        white_hot_cold=white_hc, pb_hot_cold=pb_hc,
        white_dow=white_dow, pb_dow=pb_dow,
    )
    print("  Signals: freq_recency, dirichlet, gap, positional, pair_cooccur, hot_cold, day_of_week, structural")

    # 3. Print top numbers by each signal
    print("\n[3/7] Top numbers per signal dimension:")
    top_n = 10
    print(f"  Recency-weighted top whites: {np.argsort(-white_rec)[1:top_n+1].tolist()}")
    print(f"  Dirichlet posterior top whites: {np.argsort(-white_dir)[1:top_n+1].tolist()}")
    print(f"  Most overdue whites (gap): {np.argsort(-white_gap)[1:top_n+1].tolist()}")
    print(f"  Hottest whites (recent): {np.argsort(-white_hc)[1:top_n+1].tolist()}")
    print(f"  Recency top PB: {np.argsort(-pb_rec)[1:6].tolist()}")
    print(f"  Most overdue PB: {np.argsort(-pb_gap)[1:6].tolist()}")
    print(f"  Hottest PB: {np.argsort(-pb_hc)[1:6].tolist()}")

    # 4. Generate candidates
    print(f"\n[4/7] Generating 300,000 candidate tickets (60% guided, 40% random)...")
    whites_arr, pb_arr = build_candidate_pool(
        rng=rng, n_candidates=300_000, signals=signals,
        guided_frac=0.6, temperatures=[0.5, 1.0, 2.0],
    )
    print(f"  Pool generated: {len(pb_arr)} tickets")

    # 5. Rank and pick top 5
    print(f"\n[5/7] Ensemble ranking (8-dimensional z-score)...")
    top_tickets = pick_best_tickets(
        whites_arr, pb_arr, signals,
        weights=DEFAULT_WEIGHTS, top_k=5,
    )

    print("\n" + "=" * 70)
    print("TOP 5 PICKS (diversified)")
    print("=" * 70)
    for t in top_tickets:
        w_str = " ".join(f"{x:2d}" for x in sorted(t["whites"]))
        print(f"\n  #{t['rank']}  Whites: [{w_str}]  PB: {t['powerball']:2d}")
        print(f"       Combined score: {t['combined_score']:.4f} | Structural: {t['structural']:.0%}")
        zs = t["signal_z_scores"]
        parts = [f"{k}={v:+.2f}" for k, v in sorted(zs.items(), key=lambda x: -abs(x[1]))]
        print(f"       Z-scores: {', '.join(parts)}")

    # 6. Evaluate #1 pick on held-out test set
    best = top_tickets[0]
    best_w = np.array(sorted(best["whites"]))
    best_pb = best["powerball"]

    print(f"\n[6/7] Evaluating #1 pick on {len(test)} held-out draws (10K baseline sims)...")
    eval_res = evaluate_ticket(rng, best_w, best_pb, test, n_baseline=10_000)

    print("\n" + "=" * 70)
    print("HONEST EVALUATION — HELD-OUT TEST PERIOD")
    print("=" * 70)
    print(f"  Test draws: {eval_res['n_test_draws']}")
    print(f"  White match distribution: {eval_res['white_match_dist']}")
    print(f"  Total white matches: {eval_res['total_white_matches']} (baseline mean: {eval_res['baseline']['mean_white_matches']:.1f} ± {eval_res['baseline']['std_white_matches']:.1f})")
    print(f"  PB matches: {eval_res['pb_matches']} (baseline mean: {eval_res['baseline']['mean_pb_matches']:.2f})")
    print(f"  p(random ≥ our whites): {eval_res['baseline']['p_white_ge']:.3f}")
    print(f"  p(random ≥ our PB):     {eval_res['baseline']['p_pb_ge']:.3f}")
    if eval_res["prizes_won"]:
        print(f"  Prizes won: {eval_res['prizes_won']}")
    else:
        print(f"  Prizes won: none (typical)")

    # 7. Final interpretation
    print("\n" + "=" * 70)
    print("INTERPRETATION")
    print("=" * 70)
    pw = eval_res["baseline"]["p_white_ge"]
    ppb = eval_res["baseline"]["p_pb_ge"]
    if pw < 0.05:
        print("  ✓ White ball matching is BETTER than random at 5% significance.")
    elif pw < 0.20:
        print("  ~ White ball matching shows a slight edge, but not statistically significant.")
    else:
        print("  ✗ White ball matching is NOT better than random (expected for lottery).")

    if ppb < 0.05:
        print("  ✓ Powerball matching is BETTER than random at 5% significance.")
    else:
        print("  ✗ Powerball matching is NOT better than random (expected for lottery).")

    print(f"\n  Runtime: {time.time() - t0:.1f}s")
    print("\n  REMINDER: The Powerball jackpot odds are 1 in 292,201,338.")
    print("  No statistical method can overcome the house edge.")
    print("  This analysis picks numbers that are structurally sound and")
    print("  historically favored — but every ticket has the same tiny chance.")

    return top_tickets, eval_res


if __name__ == "__main__":
    results, evaluation = main(seed=42)
```
